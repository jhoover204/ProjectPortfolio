## Task 1: Reading images into NumPy arrays [3 marks]

The company takes a picture of each clock, and saves it as a PNG image of 101x101 pixels. The folder `clock_images` contains the photos of all the clocks you need to control today.

In a PNG colour image, the colour of each pixel can be represented by 3 numbers between 0 and 1, indicating respectively the amount of **red**, the amount of **green**, and the amount of **blue** needed to make this colour. This is why we refer to colour images as **RGB** images.

- If all 3 values are 0, the pixel is black.
- If all 3 values are 1, the pixel is white.
- If all 3 values are the same, the pixel is grey. The smaller the values, the darker it is.
- Different amounts of red, green, and blue correspond to different colours.

For example, select a few colours [using this tool](https://doc.instantreality.org/tools/color_calculator/), and check the RGB values for that colour in the *RGB Normalized decimal* box. You should see that, for instance, to make yellow, we need a high value of red, a high value of green, and a low value of blue.

If you'd like more information, [this page](https://web.stanford.edu/class/cs101/image-1-introduction.html) presents a good summary about RGB images.

## Task 2: Clean up the images to extract data [6 marks]

Later in Task 3, we will use **linear regression** to find the exact position of both clock hands. To perform linear regression, we will need the **coordinates of the pixels** belonging to each hand; then, we will be able to fit a line through these pixels.

This task is concerned with extracting the correct pixel coordinates from the image.

---

ðŸš© Write a function `get_clock_hands(clock_RGB)`, which takes one input argument `clock_RGB`, a NumPy array of size 101x101x3 representing an RGB image of a clock, and returns 2 NumPy arrays with 2 columns each, such that:

- In the first array, each row corresponds to the `[row, column]` index of a pixel belonging to the **hour hand**.
- In the second array, each row corresponds to the `[row, column]` index of a pixel belonging the **minute hand**.

The goal is to obtain, for each hand, a collection of `[row, column]` coordinates which indicate where on the picture is the clock hand. You will need to figure out a way to decide whether a given pixel belongs to the hour hand, the minute hand, or neither.


---

***Important note:*** the pictures all contain some amount of noise and blur. Depending on how you decide to count a certain pixel or not as part of a clock hand, your function will pick up different pixels. There isn't just one possible set of pixel coordinates to pick up for a given image -- the most important thing is that the pixels you extract **only** belong to one of the two hands, and not to the background for example. This will ensure that you can use linear regression efficiently.

---

âœ… *Testing:* For example, for the tiny 7x7 clock below (a 7x7 pixel image is available in the `testing` folder for you to try your function):

| Clock | Hour hand | Minute hand |
|:--:|:--:|:--:|
| <img src="graphics/task2.png" alt="Task 2 example" style="width: 100px;"/> | [[1, 1]<br/> [2, 2]] | [[3, 3]<br/> [4, 3]<br/> [4, 4]<br/> [5, 4]<br/> [6, 5]] |

````python
def get_clock_hands(clock_RGB):
    """
    Goal: Find the Hour (Red) indices and the Minute (Green) indices within the supplied 3D RGB image array
    
    Input: 3D array with RGB pixel data in each layer of the array (0: Red, 1: Green, 2: Blue)
    
    Output: Two 2D arrays with indices of pixels in the Hour (red pixels) and Minute (green pixels) hands
            such that the rows of each array correspond to the [row, column] of the given hand in the image
            where a row is the y-axis coordinate of a pixel and the column is the x-axis coordinate of a pixel.
            The first array is the Hour indices and second is the Minute indices
    
    Method:
    Going to call Hour (Red) if:
        A) Red is 20% or more intense than green (ie Red/Green values > 1.2) AND
        B) Red is 20% or more intense than blue (ie Red/Blue values > 1.2)
    Going to call Minute (Green) if:
        A) Green is 20% or more intense than red (ie Green/Red values > 1.2) AND
        B) Green is 20% or more intense than blue (ie Green/Blue values > 1.2)
        
    This method is chosen since Task 1 demonstrated that Hour hand is high in red and low in others
    and Minute hand is high in green and low in others. I chose 20% or higher as a moderate cutoff
    so no pixels outside of the hand location are chosen, but enough are chosen to perform linear regression.
    Therefore, 20% is somewhat arbitrary but it is a cutoff that generates accurate interpretations
    (based on results from tasks 5-7)
    """
    ##import necessary packages within the function so the functions can be used independent of the notebook
    import numpy as np
        
    ##Implementation Strategy:

    ##To find the pixels that are 20% higher in the color of interest than other colors for:
    
    ##A) Hour:
        ##divide red_mat/green_mat and red_mat/blue_mat
        ##For each ratio matrix, check if the ratio is >= 1.2 (20% more color in red than others)
        ##Multiply both ratio matrices by each other to check that both conditions are met 
        ##(True * True = 1 * 1 = True; True * False = 1 * 0 = False; False * False = 0 * 0 = False)
        ##Check which indices meet the conditions above with np.where()
        ##And return indices of those that match the test
        
    ##B) Minute:
        ##Divide green_mat/red_mat and green_mat/blue_mat
        ##The rest of process is same as for the Hour hand.
        
    ##Find Indices 
    
    ##Hour test: red [0 dim] 20% more than both green [1] and blue [2]
    red = ((clock_RGB[:,:,0]/clock_RGB[:,:,1]) >= 1.2) * ((clock_RGB[:,:,0]/clock_RGB[:,:,2]) >= 1.2) 
    hour = np.where(red == True) ##indices where red is 20% more than both green and blue
    
    ##Minte test: green [1 dim] 20% more than both red [0] and blue [2]
    green = ((clock_RGB[:,:,1]/clock_RGB[:,:,0]) >= 1.2) * ((clock_RGB[:,:,1]/clock_RGB[:,:,2]) >= 1.2) 
    minute = np.where(green == True) ## indices where green is 20% more than both red and blue
    
    ##Note:
    #np.where() on a 2D array returns two 1D arrays, so to get a 2D array just cast with np.array
    #This yields an array with two rows where the first row is the rows in original clock_RGB
    #and second row is the columns of original clock_RGB
    #Taking the transpose will put into the format where each row is [row, column] from the original image array
    
    return np.array(hour).T, np.array(minute).T 

````

## Task 3: Calculate the angle of the two hands [9 marks]

Now that we have pixel locations for each hand, we can estimate the **angle** between each hand and the 12 o'clock position. We will use this angle later to determine the time indicated by each hand. For instance, the figure below shows the angle made by the hour hand with the 12 o'clock position.

![Angle between hour hand and 12 o'clock](graphics/angle.png)

---

ðŸš© Write a function `get_angle(coords)` which takes one input argument, a NumPy array with 2 columns representing `[row, column]` pixel coordinates of one clock hand, exactly like one of the arrays returned by `get_clock_hands()` from Task 2.

- Your function should use these pixel coordinates to find a **line of best fit** using linear regression.
- Then, using this line of best fit, you should determine and **return** the angle between the clock hand and the 12 o'clock position, measured in **radians**.

The angle should take a value between $0$ (inclusive) and $2\pi$ (exclusive) radians, where $0\, \text{rad}$ corresponds to the 12 o'clock position.

---

***Notes:***

- When performing linear regression, you will need to pay particular attention to the case where the clock hand is vertical or almost vertical.
- Beware of the correspondance between `[row, column]` index and `(x, y)` coordinate for a given pixel.
- Note that the meeting point of the 2 clock hands may not be exactly at `[50, 50]`. Some of the pictures have a small offset.
- Partial attempts will receive partial marks. For instance, if you are struggling with using linear regression, or if you don't know how to account for possible offset of the centre, you may receive partial marks if you use a simpler (but perhaps less accurate) method.

---

âœ… *Testing:* the files `task3_hourhand.txt` and `task3_minutehand.txt` are provided for you to test your function in the `testing` folder. Use `np.loadtxt()` to read them.

With these coordinates, you should find an angle of approximately 4.2 radians for the hour hand, and 5.7 radians for the minute hand.

````python
def get_angle(coords):  
    """
    Goal: Given a 2D array with [row, column] indices of the pixels that correspond to a clock hand in
          an image array, identify the angle between each hand and the 12 O'clock position using linear regression
    
    Input:
    1) coords: 2D array with coordinates of a clock hand where y-coordinates 
               are in the first column and x-coordinates in the second
    
    Output:
    2) Angle theta between the 12 O'clock position and the hand supplied
    
    Method:
    
    First perform a linear regression on the pixel points.
    Using the slope, define a vector, b, parallel to the clock hand as b = [1, slope].
    Then define a vector, a, parallel to the 12 O'clock position as a = [0, -1]
    (Use -1 as the y-component for the 12 O'clock position because the clock is plotted with a reversed y-axis)
    
    Next, calculate the angle between those two vectors to get the angle between the slope and the 12 O'clock
    by taking arcos(cos(theta)) where cos(theta) is found using the dot product formulation of cos(theta)
    ie: arcos(cos(theta) = arcos((a dot b ) / (||a|| *  ||b||))) = theta
    Since a is parallel to the 12 O'clock position and b is parallel to the hand, 
    the angle between them will also be the angle of the hand.

    This method means we don't have to worry about the center of the clock hand
    Because any vector parallel to the y-axis will yield the same angle when dotted with
    a vector on our regression line. 

    However, a hand can either be facing "down" or "up" on the regression line.
    To account for this, note that either the point with the min or max x value 
    on the hand (from the supplied array) will be closest to the center (50,50).
    When the hand points from 12-6 the point with the min x value (furthest left) is closest to the center
    When the hand points from 6 to 12 the point with the max x value (furthest right) is closest to the center
            
    Also remember that images are reversed on the y-axis, 
    so a positive slope is a hand facing down. Considering all of this:
        A) min x value is closer to center (50,50) than max x value  --> Points to 12-6. Theta = calculated theta above
        B) max x value is closer to center (50,50) than min x value --> Points to 6-12. Theta = calculated theta + pi

    Last thing to consider is when hands face either 12 or 6
    For 12 and 6, slope will be asymptotic, so we can't use the method above.
    To check for this case, I divide the standard deviation of the y-value by
    the standard deviation of the x-value. If the line is vertical, the std_dev of x will be
    close to 0, so the division yields an error. If we test for that error, we can identify 
    vertical lines.
    
    We can then find the direction of the hand in a similar way to above
    by checking if the point with the max y-value or min y-value is closer to the center
    (as opposed to max/min x-value before). Again remember y-axis is reversed in the images:
        A) Min y-value is closer to center (50,50) than max y-value --> Points to 6. Theta = Pi
        B) Max y-value is closer to center (50,50) than min y-value Theta = 0
        
    For all points close to asymptotic, but not asymptotic (ie not dividing by zero), we will still get a regression
    from the above method because the std_dev of x is not small enough the cause a ZeroDivision error when the bounds of the
    plot are (0, 101) for both x and y.
    """
    
    ##import necessary packages within the function so the functions can be used independent of the notebooks
    import numpy as np
    import scipy.stats as stats


    ##1) Test if line is vertical (or close) 
    
    ##std dev of x should be close to 0 if vertical, which will yield a division error 
    ##converting to regular float because np.floats give an error with zero division that messes
    ##up the try statement below.
    
    x_std = float(np.std(coords[:,1]))
    y_std = float(np.std(coords[:,0]))
    
    try:##testing to see if we get zero deviation in x (vertical line)
        y_std/x_std ##if not zero it will not yield an error and can run the remaining code
        vertical = False ##Test case to determine which code branch to use

        
    except ZeroDivisionError:##if the slope is a vertical line y_std/x_std should be y_std/~0 which will yield an error
        vertical = True ##if there is a division error (vertical line), we know to use the vertical line code branch
        
        
        
    ##2) Get approximate center of plot needed for directionality tests for both vertical and non-vertical hands
    approx_center = np.array([[50,50]]) ##center of plot is close to (50,50)
    
    
    
    ##3) Run a linear regression if not vertical. Identify the angle between the slope and the 12 hand 
    ##   using the dotproduct method descripe in docstring. Then modify the calculated theta value based on hand direction
    
    
    if vertical == False:

        # create linear regression object of arrays using rows in the image (zeroth column in supplied array) as y value
        regr = stats.linregress(x = coords[:,1], y = coords[:,0])

        
        ##Calculating theta initially (Before direction test)
        a = np.array([0,-1]) ##vertical pointing "up" (-1 y-component because y-axis is flipped). Parallel to 12 O'Clock
        b = np.array([1, regr.slope]) ##positive x-vector parallel to the regression line
        dotprod = np.dot(a,b) ##dotproduct of a and b
        norm_a = np.linalg.norm(a) ##magnitude of a
        norm_b = np.linalg.norm(b) ##magnitude of b
        calc_theta = np.arccos(dotprod/(norm_a * norm_b)) ##yields the angle between slope and y-axis without direction


        ##Test for direction of hand using x-value since this is the non-vertical code branch (see docstring for explanation)
        x_minpoint = coords[coords[:,1] == coords[:,1].min()] ##Pick a point with lowest x-value
        x_maxpoint = coords[coords[:,1] == coords[:,1].max()] ##point a point with highest x-value

        min_center_dist = np.linalg.norm(x_minpoint - approx_center) ##euclidean distance from approximate center
        max_center_dist = np.linalg.norm(x_maxpoint - approx_center) ##euclidean distance from approximate center


        ##Hand Directionality test for non-vertical (see docstring for explanation of logic)
        if (max_center_dist > min_center_dist): ##12 to 6 on clock
            theta = calc_theta
        else: ##6 to 12 on clock
            theta = calc_theta + np.pi
     
    
    ##4 If the hand is vertical, set angle either to pi or 0 depending on hand direction
    
    else: ##if the slope is not a vertical line or close to a vertical line (vertical=True)

        y_minpoint = coords[coords[:,0] == coords[:,0].min()] ##pick a point with lowest y-value (as opposed to x above)
        y_maxpoint = coords[coords[:,0] == coords[:,0].max()] ##pick a point with highest y-value (as opposed to x above)

        min_center_dist = np.linalg.norm(y_minpoint - approx_center) ##euclidean distance from approximate center
        max_center_dist = np.linalg.norm(y_maxpoint - approx_center) ##euclidean distance from approximate center

        ##Hand directionality test for non-vertical; here is checking y-val distances from center
        if (max_center_dist > min_center_dist): ##6 on clock
            theta = np.pi
        else: ##12 on clock
            theta = 0
            
    return(theta)

##Testing with supplied data
import numpy as np
hour = np.loadtxt(fname = "./testing/task3_hourhand.txt")
minute = np.loadtxt(fname = "./testing/task3_minutehand.txt")

print(get_angle(hour))
print(get_angle(minute))
````

## Task 4: Visualising the clock [6 marks]

ðŸš©  Use `matplotlib` and your artistic skills to visualise the clock. Write a function `draw_clock(angle_hour, angle_minute)` that takes 2 input arguments, corresponding to the two angles of the clock hands, and draw a clock with the precise location of both hands.

âœ… *Testing:* with `angle_hour` set to $\frac{\pi}{3}$ and `angle_minute` set to $\frac{11\pi}{6}$, the hour hand should point exactly at 2, and the minute hand should point exactly at 11.

There is also an example image in the `testing` folder, which was produced entirely with `matplotlib`. This is just to give you an idea of what is possible to do -- you shouldn't attempt to reproduce this particular example, don't hesitate to get creative!

````python
##Idea for a polar plot came from the docs:
##Line 63
##https://matplotlib.org/stable/gallery/pie_and_polar_charts/polar_demo.html
##Accessed 06 Nov, 2021

##Changing location of the 0 radian spot on a polar plot 
##and removing tick labels came from this stack exchange post
##Lines 69-71; Author: Julian Spronck
##https://stackoverflow.com/questions/40852861/hide-radial-tick-labels-matplotlib
##Accessed 06 Nov, 2021

##Idea for using annotate to draw arrows came from the docs:
##Lines 90-96
##https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html
##Accessed 06 Nov, 2021


def draw_clock(angle_hour, angle_minute):
    """
    Goal: Draw a clock with hands at the angles specified in the argument
    
    Input: angle hour (radians), angle_minute(radians)
    
    Output: Plot with hands at the angles specified in the arguments
    
    Method:
    
    Plot the supplied angles and hands using a polar plot 
    that has been modified so that 0 radian point is at the 12 O'Clock position.
    """
    
    ##import necessary packages within the function so the functions can be used independent of the notebooks
    import numpy as np
    import matplotlib.pyplot as plt
    
    
    ##1) Pre-specify data for the plots
    
    ##specify circumference and radius
    radius = 45 ##chosen somewhat arbitrarily; having a number keeps things consistent
    circ = 2 * np.pi * radius

    
    ##arrow data for plotting later; (x, y) format for a polar plot is (radian, radius)
    hourhand_loc = (angle_hour, 4/9*radius) ##this ratio (4/9) gives a good size for the hour hand in the image
    minhand_loc = (angle_minute, 8/9*radius) ##this ratio (8/9) gives a good size for the minute hand in the image

    
    ##tick data for plotting later
    tick_loc = np.linspace(0, 2 * np.pi, 12*5, endpoint = False) ##location of hour and minute ticks not including 2pi 
    hour_lab  = np.linspace(0, 11, 12).astype(int).astype(str) ##labels of hour ticks in character format
    hour_lab[0] = 12 ##relabel top of clock 12 not 0
    tick_labels = ["â€¢"] * 60 ##bullets as tick labels for every minute; hours will replace some in next line
    tick_labels[::5] = hour_lab ##replace bullet points with the hour labels at every 5th position
    
    
    ##2) make plots in polar coordinate system
    
    ##This line was adapted from docs. The subplot_kw argument specifies that we
    ##will plot with a keyword; in this case the keyword is 'projection', and we are
    ##specifying the type of projection as 'polar' to get a polar coordinate system projection
    
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize = (8,8))

    
    ##Change theta zero location and get rid of axis ticks
    ##This code was adapted from a stack post. See notes before function.
    
    ax.set_theta_zero_location('N')##sets zero to "north" or top of plot; currently counterclockwise
    ax.set_theta_direction(-1)##reverses direction so it is clockwise
    ax.set_yticklabels([]) ##get rid of y axis ticks (radius labels)

    
    ##Set x axis ticks
    ax.set_xticks(tick_loc)##60 equally spaced locations across the circumference
    ax.set_xticklabels(tick_labels)##label names with number for the hours and bullets for the minutes
    ax.tick_params(axis = "x", direction= "in", pad = -20, labelsize = 16) ##move them inside the clock

    
    ##Plot hands as arrow
    ##This code idea came from the docs (see top of notes)
    ##ax.annotate adds text to a plot. Here we've added "" for empty text
    ##It then adds an arrow from xytext (here center point (0,0)) to the
    ##location specifed in xy (here we predefined the location as the angles
    ##of the hand for x and radius for y). The arrowprops define the arrow
    ##properties in a dictionary. The arguments are somewhat self-explanatory. 
    ##zorder gives which plot layer to place the object in (0 at bottom/first layer)
    
    ##hour
    ax.annotate("", xytext = (0, 0) , xy = hourhand_loc, ##tail located at center; head at angle and radius specified above
                arrowprops = dict(headwidth = 8, edgecolor = 'k', facecolor = 'r', ##edge black, face green
                                  lw = 1), zorder = 0) ##aesthetics, places hour at bottom layer; lw = linewidth
    ##minute
    ax.annotate("", xytext= (0, 0), xy = minhand_loc, ##tail located at center; head at angle and radius specified above
                arrowprops= dict(headwidth = 8, edgecolor = 'k', facecolor = 'g', ##edge black, face green
                                  lw = 1), zorder = 1) ##aesthetics, places minute at next layer; lw = linewidth

    
    ##Add center circle for aesthetics at x,y = (0,0)
    ax.scatter(0, 0 , c = "k", zorder = 3) ##make zorder higher than the rest so it is on top

    
    ##Plot
    ax.plot(circ, radius)##creating polar plot with preselected circumference and radius
    ax.grid(False) ##get rid of grid
    ax.set_title("Fig 1. Polar plot of clock. Hand location is specified in radians with 12 as 0 radians.", y = -0.1)
    plt.show()
   

##Testing with supplied test data
import matplotlib.pyplot as plt
import numpy as np

angle_hour = np.pi/3
angle_minute = 11*np.pi/6
draw_clock(angle_hour, angle_minute)
````

````python
def analog_to_digital(angle_hour, angle_minute):
    """
    Goal: Turn the angles (in radians) of the hour hand and minute hand into a digital clock with time in hh:mm format
    
    Input: angle hour (radians), angle_minute(radians)
    
    Output: A string representing a digital clock with time in hh:mm format
    
    Method:
    
    Taking a clock as the unit circle, the clock can be broken up into 12 intervals (hours) 
    of length pi/6, or 60 intervals (minute) of pi/30. Using these interval widths, 
    we can identify how many intervals a given hour or minute hand has surpassed, 
    and use this to find the hour and minute. 
    
    Note: For hours, the 0th interval is in the 12 O'clock hour.
    
    For example:
    
    An hour angle of pi/4 is at (pi/4)/(pi/6) intervals or 3/2 (1 and 1/2) intervals, so it is in the first hour
    A minute hand of 2pi/7 is at (2pi/7)/(pi/30) intervals or 60/7 (8 and 4/7) intervals, so it is in the 8th minute
    """
    
    ##import necessary packages within the function so the functions can be used independent of the notebooks
    import numpy as np
    
    ##setting interval widths
    hour_intwid = np.pi/6 ##equiv to 2pi/12
    min_intwid = np.pi/30 ##equiv to 2pi/60
    
    ##Identifying interval (as integer to avoid decimals in output)
    hour_int = int(np.floor(angle_hour/hour_intwid)) ##finds the hour interval, ignoring how far into the interv it has gone
    min_int = int(np.floor(angle_minute/min_intwid)) ##finds the min interval, ignoring how far into the interv it has gone
    
    if hour_int == 0: hour_int = 12 ##convert the hour to 12 if in the 0th interval
    
    ##converting to string output
    
    ##Hour formatting; coerce to string regardless, if less than 10 add leading 0
    hour_int = "0" + str(hour_int) if hour_int < 10 else str(hour_int) 
        
    ##Minute formatting; coerce to string regardless, if less than 10 add leading 0
    min_int = "0" + str(min_int) if min_int < 10 else str(min_int) 

    return(hour_int + ":" + min_int)##add all strings together and return


##Testing with supplied data
import numpy as np

angle_hour = np.pi/3
angle_minute = 11*np.pi/6

print(analog_to_digital(angle_hour, angle_minute))
````

## Task 6: Find the misalignment [5 marks]

Now that you have extracted useful information from the pictures, you need to check if the two hands are aligned properly. To do so, you will need to find the exact time that the **small hand** is showing, in hours and minutes. Then, compare with the minutes that the big hand is showing, and report the difference.

Note that the misalignment will never be more than 30 minutes. For example, if you read a 45-minute difference between the minutes indicated by the hour hand and by the minute hand, you can realign the minute hand by 15 minutes in the other direction instead.

ðŸš© Write a function `check_alignment(angle_hour, angle_minute)` which returns the misalignment in minutes.

Make sure you structure you code sensibly. You may wish to use some intermediate functions to do the sub-tasks.

````python
##Method to split numbers into the integer and decimal component came from this:
##Line 70; Authors: mhyfritz, Mark Dickinson
##https://stackoverflow.com/questions/6681743/splitting-a-number-into-the-integer-and-decimal-parts
##Accessed 06 Nov, 2021

def check_alignment(angle_hour, angle_minute):
    """
    Goal: Check the aligment between the minute hand and the hour hand to see
          if the minute hand needs adjustment and, if yes, by how much.
    
    Input: angle hour (radians), angle_minute(radians)
    
    Output: Absolute misalignment in minutes
    
    Method:
    
    Taking a clock as the unit circle, the clock can be broken up into 12 intervals (hours) 
    of length pi/6, or 60 intervals (minutes) of pi/30. Using these interval widths, 
    we can identify how many intervals a given hour or minute hand has surpassed. 
    We can then use this to find the hour and minute for both the hour and the minute hands.
    
    For example: 
    
    A) An hour angle of pi/4 is at (pi/4)/(pi/6) intervals or 3/2 (1 and 1/2) intervals, 
       so it is in the first hour and has gone halfway towards the next hour (30 minutes)
       
    B) A minute hand of 2pi/7 is at (2pi/7)/(pi/30) intervals or 60/7 (8 and 4/7) intervals, 
       so it is in the 8th minute. The remaining 4/7 is the seconds, which we are not concerned with.
    
    For the hour hand, the whole number will be the hour (0 is the 12th hour) and the decimal
    is how far into the next hour (in minutes) it is. So multiplying the decimal component 
    by 60 will give the minutes. Any decimal after this represents seconds.
    
    For the minute hand, the whole number is the minute and the decimal is the seconds.
    Again, we are not concerned with seconds.
    
    To find the misalignment, just subtract the hour hand minutes and the minute hand 
    minutes, take the absolute value, and round the result to the nearest number since 
    we are not interested in seconds.
    
    Since the max difference is 30 minutes, any value over 30 should be subtracted from 60.
    (A 45 minute difference can be corrected by moving the minute hand 15 minutes 
    (60 - 45) in the other direction)

    Notes: Since we are only reporting misalignment in minutes, 
           I am ignoring the amount of time in hours that the hour hand shows. 
          
           Additionally, the reported misalignment only gives the absolute misalignment. 
           The direction of misalignment could easily be found by subtracting any misalignments 
           over 30 from 60 and multiplying the result by negative 1. 
    """
    
    ##import necessary packages within the function so the functions can be used independent of the notebooks
    import numpy as np
    from math import modf ##for splitting a decimal into whole number and decimal
    
    ##setting up interval widths
    hour_intwid = np.pi/6
    min_intwid = np.pi/30
    
    ##To identify the hour hand minutes we find how many intervals the hour hand has surpassed,
    ##by dividing the angle of the hand by the amount of radians in an hour interval (specified above).
    ##We then subset the decimal component to find how far into the minutes it 
    ##has gone. modf (function found in stack post; see notes above function) 
    ##splits a float into a tuple with the first element as the decimal 
    #and the second element as the whole number. So modf will supply the info we need.
    ##ie modf(1.031) yields (0.031, 1.0), so subsetting modf(<hours>.<proportionintonexthour>)[0] 
    ##gives the proportion into the next hour (the [1] slot would give how many hours, but not relevant here)
    
    hour_mins = modf(angle_hour/hour_intwid)[0] ##yields the proportion of the next hour the hour hand has gone into
    hour_mins = hour_mins * 60 ##convert proportion of the hour into minutes (remaining decimal is seconds)
    
    ##Identifying minute hand minutes
    min_mins = angle_minute/min_intwid ##gives how many minute intervals have passed (remaining decimal is seconds)
    
    ##Calculate the misalignment and round to get rid of seconds
    misalignment = int(round(np.linalg.norm(hour_mins - min_mins), 0)) ##absolute difference in minutes; return as int
    
    ##correct for misalignments greater than 30 minutes, by subtracting from 60 minutes
    if misalignment > 30: misalignment = 60 - misalignment
        
    return(misalignment) 

##Testing with data supplied in task 4
import numpy as np

angle_hour = np.pi/3
angle_minute = 11*np.pi/6

print((check_alignment(angle_hour, angle_minute)))
````

````python
##How to remove milliseconds from datetime output came from:
##Line 109; Author: Sven Marnach
##https://stackoverflow.com/questions/7999935/python-datetime-to-string-without-microsecond-component
##Accessed 7 Nov, 2021

##Listing files in a directory came from:
##Line 120; Authors: Yul Kang, pycruft 
##https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory
##Accessed 7 Nov, 2021

##Reordering a list of sublists by one of the sublist indices was found here:
##Line 156
##https://docs.python.org/3/howto/sorting.html
##Acessed 7 Nov, 2021

##How to structure a string with alignment and padding so that output is vertically aligned was found at:
##Line 199
##https://docs.python.org/3/library/string.html#formatspec
##Accessed 7 Nov, 2021

def validate_clock(filename):
    """
    Goal: Take an RGB image of a clock and report the misalignment between the hour hand and the minute hand
    
    Input:
    filename: path to file with clock rgb image
    
    Output:
    Misalignment of hands in minutes
    
    Implementation:
    Read in the data as an image file using matplotlib.pyplot.imread. 
    Then, using 3 prewritten functions, get_clock_hands, get_angle, and check_alignment 
    (see help docs for those functions), report the misalignment.
    
    Misalignment will be no more than 30 minutes as anything over 30 can be fixed by realigning the hand in the
    counter-clockwise direction.
    """
    
    ##import necessary packages within the function so the functions can be used independent of the notebooks
    import matplotlib.pyplot as plt
    ##numpy is imported within the other functions so we don't need to import
    
    ##read in image as 3D array
    image_array = plt.imread(fname = filename) ##3D array with red, green, and blue in the 0, 1, and 2 z-order, respectively
    
    ##get location of clock hands within the image and return as tuple of 2 arrays
    ##with hour in the 0 slot and minute in the 1 slot.
    ##format of the arrays is [row, column] where row is the y-axis location within the image array
    ##and column is the x-axis location within the image array
    clock_hands = get_clock_hands(image_array)
    
    ##get angles of both hands and report as a tuple with hour angle in the 0 slot and minute in the 1 slot
    angles = (get_angle(clock_hands[0]), get_angle(clock_hands[1])) ##[hour angle, minute angle]
    
    ##check the alignment between the two misalignment between the hands
    misalignment = check_alignment(angles[0], angles[1])##check_alignment(hour angle, minute angle)
    
    return(misalignment)


def validate_batch(path, tolerance):
    """
    Goal: Generate a QC .txt file for a batch of clock images within a supplied directory. The file will contain  
          the batch information, the analysis time, the number of clocks analysed, the number of 
          clocks passing QC, the percentage passing QC, and a list of clocks that did not pass
          with their corresponding misalignments. The .txt file will be stored within the parent directory
          of the supplied path. ie if path = "data/clock_images/batch_0/" the QC report for batch 0 will be
          in a file in a new direcotory: "data/clock_images/QC_Reports/"
    
    Input:
    A) path: path to directory with clock images
    B) tolerance: maximum number of minutes a clock can be misaligned before needing to report in the batch QC
    
    Output:
    A .txt file called batch_X_QC.txt that is stored in the parent directory of the supplied path in a 
    new directory called "QC_Reports" (ie "../QC_reports/batch_X_QC.txt") which contains:
        A) Batch number
        B) Time of QC analysis
        C) Total number of clocks analyzed
        D) Number of clocks that passed QC
        E) Percent Passing
        F) List of clocks that did not pass QC along with their misalignment (Not including direction of misalignment)
        
    Implementation:
    
    The function first gets a list of files in the clock image directory. It then 
    performs the validate_clock function on each file in the directory. From there, 
    it tests if the clock passes the tolerance specifies. If it does, it counts
    that clock as passing, if it does not, it logs the clocks information along with 
    the amount it is misaligned.
    
    At the end, it summarizes all of this data in a .txt file and places into a new folder called
    "QC_Reports" in the parent directory of the supplied path.
    """
    
    ##import necessary packages within the function so the functions can be used independent of the notebooks
    from os import listdir, mkdir ##used for checking filenames, manipulating the paths, and creating directories
    from os import path as ospath ##need to specify this new name since the function argument is path
    from datetime import datetime ##used for getting time of analysis
    
    
    ##1) Generate and organize Data
    
    
    ##normalize path input, get batch info, and get time of analysis. time was adapted from a stack post
    
    path = ospath.normpath(path) ##gets the supplied path into a normalized format (removes the trailing /)
    batch = ospath.basename(path).strip("batch_") ##gets the lowest directory (ie the batch name) and removes "batch_" 
    time = datetime.now().strftime('%Y-%m-%d, %H:%M:%S') ##gets current time in format YYYY-MM-DD, hh:mm:ss without ms
    
    
    ##Get filenames to loop over and total number of clocks
    
    ##the list comprehension was adapted from a post on stack (see notes before function)
    ##for f in listdir(path) gets the list of files within the specified directory
    #if ospath.isfile(ospath.join(path,f)) first turns the filename into a proper path with parent directories
    ##and then tests if it is a file. If it is a file, the list comprehension adds the path/file to the list "files"
    
    files = [ospath.join(path, f) for f in listdir(path) if ospath.isfile(ospath.join(path, f))] 
    total_clocks = len(files)
    
    
    ##Initalize lists and counters for data summmary
    passing_clocks = 0 ##counter for clocks; updated in loop
    misaligned_clocks = [] ##will contain a list of lists. Each inner list will have the clock name and misalignment 
    
    
    ##loop over all clocks to gather QC data (Number passing and misaligned clock list with misalignment)
    for i in range(total_clocks):
        misalignment = validate_clock(files[i]) ##get the misalignment for each clock with predefined validate_clock fxn
        
        if misalignment <= tolerance: ##if the clock passes QC within tolerated error
            passing_clocks += 1 ##count the clocks that pass
            
        else: ##if the clock does not pass QC within tolerated error
            
            ##Create a list with the clockname and misalignment
            ##Add this list to a list that contains ALL misaligned clock data (used for QC report)
            
            misaligned_data = [ospath.basename(files[i]).strip(".png"), misalignment] ##basename gets clock and .strip removes ".png"
            misaligned_clocks.append(misaligned_data) ##add the misalignment data for a given clock to the summary file
    
    
    ##Calculate passing percentage
    passing_perc = round(passing_clocks/total_clocks * 100, 1)##more readable to calculate here than when writing to .txt
    
    
    ##Place misaligned clocks in decreasing order based on misalignment (second item in sublists)
    
    ##The method to do this was found in the docs (see notes before function)
    ##The reverse argument tells it to be in descending order; 
    ##key = lambda ... tells it to order the list based on second sublist [1] index
    ##that I have named "misaligned"
    
    misaligned_clocks.sort(reverse = True, key = lambda misaligned: misaligned[1])
    
    
    ##2) writing a text file with all summary data
    
    
    ##make new directory called "QC_reports" in the parent directory of the supplied path
    output_dir = ospath.join(ospath.dirname(path), "QC_reports") ##make name for output directory in parent directory
    if(ospath.isdir(output_dir) == False): os.mkdir(output_dir) ##make new directory if it does not exist
    
    
    ##create output file name (within QC_Reports directory)
    output_file = ospath.join(output_dir, f'batch_{batch}_QC.txt')##making name for output file in the output dir
    
    
    ##writing lines with summary data
    line1 = f'Batch number: {batch}\n'
    line2 = f'Checked on {time}\n'
    line3 = '\n'##creates spacer line for formatting
    line4 = f'Total number of clocks: {total_clocks}\n'
    line5 = f'Number of clocks passing quality control ({tolerance}-minute tolerance): {passing_clocks}\n'
    line6 = f'Batch quality: {passing_perc}%\n' 
    line7 = '\n' ##creates spacer line for formatting
    line8 = 'Clocks to send back for readjustment:\n'
    
    
    ##using "with open" to prevent need to close the file and writing to a new file in the specified output directory
    
    with open(f'{output_file}', "w+") as output:
        output.writelines([line1, line2, line3, line4, line5, line6, line7, line8]) ##write the above lines on their own line
        
        ##now use a for loop to format each line for the misaligned clock info and write each clock line separately
        for i in range(len(misaligned_clocks)): 
            
            ##Structuring a string to make a table that is vertically aligned came from the docs
            ##(see notes before function). The format below is:
            
            ##{:<12} creates a left_aligned object,and fills the first part with the clock name 
            ##from misaligned_clocks[i][0] (at most 9 characters for 'clock_XXX'), 
            ##and pads the rest until the 12th character with empty space. This ensures that the 
            ##misalignment data will all be vertically aligned at the 13th character
            ##We then add (with '+') the misaligned minutes amount and "min" with a newline character
            
            misaligned = '{:<12}'.format(f'{misaligned_clocks[i][0]}') + f'{misaligned_clocks[i][1]} min\n'
            output.write(misaligned) ##write to output file


            
##for all the batches we have (0-4) generate a QC report
for i in range(5):
    batch_folder = './clock_images/batch_' + str(i)
    validate_batch(batch_folder, tolerance = 3)
````
